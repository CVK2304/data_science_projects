Importing Required Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

Reading file

```python
file= pd.read_csv("RDA.csv")
```

Renaming Column names

```python
file.rename(columns={
    'States/UTs': 'State',
    'State/UT-Wise Total Number of Road Accidents during 2016': 'Accidents_2016',
    'State/UT-Wise Total Number of Road Accidents during 2017': 'Accidents_2017',
    'State/UT-Wise Total Number of Road Accidents during 2018': 'Accidents_2018',
    'State/UT-Wise Total Number of Road Accidents during 2019 - Numbers': 'Accidents_2019',
    'State/UT-Wise Total Number of Road Accidents during 2019 - Rank': 'Accident_Rank_2019',
    'Share of States/UTs in Total Number of Road Accidents - 2016': 'Share_2016',
    'Share of States/UTs in Total Number of Road Accidents - 2017': 'Share_2017',
    'Share of States/UTs in Total Number of Road Accidents - 2018': 'Share_2018',
    'Share of States/UTs in Total Number of Road Accidents - 2019': 'Share_2019',
    'Total Number of Accidents Per Lakh Population - 2016': 'Acc_Per_Lakh_2016',
    'Total Number of Accidents Per Lakh Population - 2017': 'Acc_Per_Lakh_2017',
    'Total Number of Accidents Per Lakh Population - 2018': 'Acc_Per_Lakh_2018',
    'Total Number of Accidents Per Lakh Population - 2019': 'Acc_Per_Lakh_2019',
    'Total Number of  Road Accidents per 10,000 Vehicles - 2016': 'Acc_Per_10k_Vehicles_2016',
    'Total Number of  Road Accidents per 10,000 Vehicles - 2017': 'Acc_Per_10k_Vehicles_2017',
    'Total Number of  Road Accidents per 10,000 Vehicles - 2018': 'Acc_Per_10k_Vehicles_2018',
    'Total Number of  Road Accidents per 10,000 Vehicles - 2019': 'Acc_Per_10k_Vehicles_2019',
    'Total Number of  Road Accidents per 10,000 Km of Roads - 2016': 'Acc_Per_10k_Km_2016',
    'Total Number of  Road Accidents per 10,000 Km of Roads - 2017': 'Acc_Per_10k_Km_2017',
    'Total Number of  Road Accidents per 10,000 Km of Roads - 2018(P)': 'Acc_Per_10k_Km_2018'
}, inplace=True)
```

Find Missing Values

```python
file.isnull().sum()
```

Filling the missing values with mean value

```python

numeric_cols = file.select_dtypes(include=[np.number]).columns
file[numeric_cols] = file[numeric_cols].fillna(file[numeric_cols].mean())
```

Now, again checking the missing values

```python
file.isnull().sum()
```

Performing Statistics 

```python
file.describe()
```

Performing Standard Deviations

```python
accident_cols = [col for col in file.columns if 'accident' in col.lower()]
std_values = file[accident_cols].std()
print("\nStandard Deviation of Accident Frequencies:")
std_values
```

Distribution of Accident Frequency

```python
plt.figure(figsize=(10,5))
sns.histplot(file[accident_cols[0]], bins=20, kde=True)
plt.title("Distribution of Accident Frequency")
plt.xlabel("Accident Count")
plt.ylabel("Density")
plt.show()
```

Creating a Heatmap 

```python
plt.figure(figsize=(12,8))
numeric_file = file.select_dtypes(include=[np.number])
sns.heatmap(numeric_file.corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, linecolor='black')
plt.title("Correlation Matrix of Accident Data", fontsize=14)
plt.xticks(fontsize=10, rotation=45)
plt.yticks(fontsize=10)
plt.show()

```

Visualizing Barplot before using recommendations

```python
if 'State' in file.columns and accident_cols:
    state_accidents = file.groupby('State')[accident_cols[0]].sum().sort_values(ascending=False)
    state_accidents = state_accidents[~state_accidents.index.str.contains('Total', case=False, na=False)]
    highest_accident_state = state_accidents.idxmax()
    
    plt.figure(figsize=(14, 7))
    colors = sns.color_palette("Reds", len(state_accidents))
    colors.reverse()  # Darker shades for highest values
    sns.barplot(x=state_accidents.index, y=state_accidents.values, hue=state_accidents.index, palette=colors, legend=False)
    plt.xticks(rotation=90, fontsize=10)
    plt.yticks(fontsize=10)
    plt.title("Total Accidents by State", fontsize=14)
    plt.xlabel("State", fontsize=12)
    plt.ylabel("Total Accidents", fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.show()
    print(f"\nState with highest accidents: {highest_accident_state}")
```

Using Heatmap visualize the data

```python
heatmap_data = file[['State', 'Accidents_2016', 'Accidents_2017', 'Accidents_2018', 'Accidents_2019']]
heatmap_data = heatmap_data[~heatmap_data['State'].str.contains('Total', case=False, na=False)]
heatmap_data.set_index('State', inplace=True)
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, cmap="Reds", annot=True, fmt=".0f", linewidths=0.5)
plt.title("Road Accidents in Different States (2016-2019)", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("State", fontsize=12)
plt.show()
```

Consider the state Tamil Nadu, Visualize the road accidents before and after using line plot

```python
state_to_visualize = "Tamil Nadu"  # Change state as needed
if state_to_visualize in file["State"].values:
    state_data = file[file["State"] == state_to_visualize]
    years = [2016, 2017, 2018, 2019]
    accident_counts = state_data[["Accidents_2016", "Accidents_2017", "Accidents_2018", "Accidents_2019"]].values.flatten()
    reduced_accidents = accident_counts * 0.85  
    
    plt.figure(figsize=(10, 5))
    plt.plot(years, accident_counts, marker='o', linestyle='-', color='red', label='Before Measures')
    plt.plot(years, reduced_accidents, marker='s', linestyle='--', color='green', label='After Measures')
    
    plt.xlabel("Year")
    plt.ylabel("Total Accidents")
    plt.title(f"Effectiveness of Safety Measures in {state_to_visualize}")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()
    
    print(f"\nAccident reduction analysis for {state_to_visualize} shown above.")
else:
    print(f"\nState '{state_to_visualize}' not found in dataset.")
```

Compare all states before and after using barplot

```python
if 'State' in file.columns and accident_cols:
    state_accidents = file.groupby('State')[accident_cols[0]].sum().sort_values(ascending=False)
    state_accidents = state_accidents[~state_accidents.index.str.contains('Total', case=False, na=False)]
    highest_accident_state = state_accidents.idxmax()
    
    reduced_accidents = state_accidents * 0.85  # Assuming a 15% reduction
    
    fig, ax = plt.subplots(figsize=(14, 7))
    
    colors_before = sns.color_palette("Reds", len(state_accidents))
    colors_after = sns.color_palette("Greens", len(state_accidents))
    
    highest_index = list(state_accidents.index).index(highest_accident_state)
    colors_before[highest_index] = (0, 0, 0.6)  # Darker blue for highest state
    colors_after[highest_index] = (0, 0.6, 0)  # Darker green for highest state
    
    states = np.repeat(state_accidents.index, 2)
    accidents = np.column_stack((state_accidents.values, reduced_accidents.values)).flatten()
    labels = np.tile(["Before", "After"], len(state_accidents))
    
    palette = [color for pair in zip(colors_before, colors_after) for color in pair] 
    
    sns.barplot(x=states, y=accidents, hue=labels, palette={"Before": "red", "After": "green"}, ax=ax)

    ax.set_xticks(range(len(state_accidents)))
    ax.set_xticklabels(state_accidents.index, rotation=90, fontsize=10)
    ax.set_title("Total Accidents by State (Before and After Recommendations)", fontsize=14)
    ax.set_xlabel("State", fontsize=12)
    ax.set_ylabel("Total Accidents", fontsize=12)
    ax.grid(axis='y', linestyle='--', alpha=0.6)
    
    plt.legend(title="Status")
    plt.tight_layout()
    plt.show()
```
